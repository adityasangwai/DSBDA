```python
labelEncoder=preprocessing.LabelEncoder()
```

```python
for col in df:
    df[col]=labelencoder.fit_transform(df[col])
```

```python
df
```

```python
df1=pd.concat([df['math score'],df['reading score'],df['writing score']],axis=1)
```

```python
df1
```

```python
sns.boxplot(data=df1)
```

```python
Q1=df1.quantile(0.25)
print("\nQ1\n",Q1)
Q3=df1.quantile(0.75)
print("\nQ2\n",Q3)
IQR=Q3-Q1
print("\nIQR\n",IQR)
```

```python
low=Q1-(1.5*IQR)
high=Q3+(1.5*IQR)
print("\nlow\n",low)
print("\nhigh\n",high)
```

```python
dfnew=df1[~((df1<low) | (df1>high)).any(axis=1)]
```

```python
dfnew
```

```python
sns.boxplot(data=dfnew)
```
