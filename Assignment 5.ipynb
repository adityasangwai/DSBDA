```python
#creating bins by randomly selecting some sample points
bins=[17,35,54,71,90]
labels=['17-34','35-53','54-70','71-90'] #label for new group
df['AgeGroup']=pd.cut(df['Age'],bins=bins,labels=labels,right=False)
```

```python
df
```

```python
df.groupby('AgeGroup')['Salary'].mean()
```

```python
df.groupby('AgeGroup')['Salary'].median()
```
